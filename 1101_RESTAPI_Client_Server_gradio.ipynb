# **3. Gradio로 Header로 서버와 클라이언트가 통신하면서 직접 DB를 변경**

```python
from fastapi import FastAPI, Header, HTTPException
from sqlalchemy import create_engine, Column, String, Integer
from sqlalchemy.orm import declarative_base, sessionmaker
import uvicorn
import nest_asyncio

# SQLAlchemy 설정
DATABASE_URL = "sqlite:///./test.db"
engine = create_engine(DATABASE_URL)
SessionLocal = sessionmaker(autocommit=False, autoflush=False, bind=engine)
Base = declarative_base()

# 데이터베이스 모델 정의
class HeaderInfo(Base):
    __tablename__ = "header_info"
    id = Column(Integer, primary_key=True, index=True)
    user_agent = Column(String, index=True)
    custom_header = Column(String, index=True)

# 데이터베이스 테이블 생성
Base.metadata.create_all(bind=engine)

# FastAPI 인스턴스 생성
app = FastAPI()

# 요청 헤더 정보를 데이터베이스에 저장하는 엔드포인트
@app.post("/store-header/")
async def store_header(user_agent: str = Header(None), custom_header: str = Header(None)):
    if not user_agent or not custom_header:
        raise HTTPException(status_code=400, detail="Missing headers")

    # 데이터베이스 세션 생성
    db = SessionLocal()
    try:
        # 헤더 정보 저장
        header_info = HeaderInfo(user_agent=user_agent, custom_header=custom_header)
        db.add(header_info)
        db.commit()
        db.refresh(header_info)
        return {"message": "Header information stored successfully", "id": header_info.id}
    except Exception as e:
        db.rollback()
        raise HTTPException(status_code=500, detail="Database error: " + str(e))
    finally:
        db.close()

# 데이터베이스에서 헤더 정보를 가져오는 엔드포인트
@app.get("/get-header/{header_id}")
async def get_header(header_id: int):
    db = SessionLocal()
    try:
        header_info = db.query(HeaderInfo).filter(HeaderInfo.id == header_id).first()
        if header_info is None:
            raise HTTPException(status_code=404, detail="Header not found")
        return {
            "id": header_info.id,
            "user_agent": header_info.user_agent,
            "custom_header": header_info.custom_header,
        }
    finally:
        db.close()

# Colab에서 FastAPI 실행을 위한 설정
def run_app():
    from google.colab import output
    output.serve_kernel_port_as_iframe(8000)
    nest_asyncio.apply()
    uvicorn.run(app, host="0.0.0.0", port=8000)

# Colab 환경에서 실행할 때 아래 함수를 호출
run_app()

```

### **Simplest Code**

```python
import sqlite3
import gradio as gr

# 데이터베이스 파일 경로 설정
DATABASE_URL = "./test.db"  # SQLite 데이터베이스 파일 경로

# 데이터베이스 초기화 - 테이블 생성
# 데이터베이스 연결을 열고 테이블이 없을 경우 생성합니다.
# 기본 키로 자동 증가하는 ID 필드
# User-Agent 정보를 저장할 필드
 # Custom Header 정보를 저장할 필드
with sqlite3.connect(DATABASE_URL) as conn:
    conn.execute('''
        CREATE TABLE IF NOT EXISTS header_info (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_agent TEXT,
            custom_header TEXT
        )
    ''')

# 요청 헤더 정보를 데이터베이스에 저장하는 함수
def store_header_function(user_agent: str, custom_header: str):
    # user_agent나 custom_header가 없을 경우 오류 메시지 반환
    if not user_agent or not custom_header:
        return "Missing headers"
    # 데이터베이스에 연결하고 데이터를 저장
    with sqlite3.connect(DATABASE_URL) as conn:
        cursor = conn.cursor()
        # SQL 쿼리를 실행하여 header 정보를 삽입합니다.
        cursor.execute(
            "INSERT INTO header_info (user_agent, custom_header) VALUES (?, ?)",
            (user_agent, custom_header),
        )
        conn.commit()  # 변경사항을 커밋하여 데이터베이스에 반영
        return f"Header information stored successfully with id: {cursor.lastrowid}"

# 데이터베이스에서 헤더 정보를 가져오는 함수
def get_header_function(header_id: int):
    # 데이터베이스에 연결하고 특정 ID의 헤더 정보를 검색
    with sqlite3.connect(DATABASE_URL) as conn:
        cursor = conn.cursor()
        # SQL 쿼리를 실행하여 ID에 해당하는 header 정보를 조회합니다.
        cursor.execute("SELECT * FROM header_info WHERE id = ?", (header_id,))
        row = cursor.fetchone()  # 조회 결과에서 한 행을 가져옵니다.
        if row is None:
            return "Header not found"  # 해당 ID의 정보가 없을 경우 메시지 반환
        # 조회된 정보를 딕셔너리 형태로 반환
        return {
            "id": row[0],
            "user_agent": row[1],
            "custom_header": row[2],
        }

# Gradio 인터페이스 정의
with gr.Blocks() as demo:
    gr.Markdown("# Header Info Storage and Retrieval")  # 웹 UI 제목 표시
    # 헤더 정보를 저장하는 탭
    with gr.Tab("Store Header"):
        user_agent_input = gr.Textbox(label="User Agent")  # User-Agent 입력 필드
        custom_header_input = gr.Textbox(label="Custom Header")  # Custom Header 입력 필드
        store_button = gr.Button("Store Header")  # 저장 버튼
        store_output = gr.Textbox(label="Output")  # 저장 결과를 표시할 출력 필드
        # 버튼 클릭 시 store_header_function 호출
        store_button.click(store_header_function, inputs=[user_agent_input, custom_header_input], outputs=store_output)
    # 헤더 정보를 조회하는 탭
    with gr.Tab("Get Header"):
        header_id_input = gr.Number(label="Header ID")  # 헤더 ID 입력 필드
        get_button = gr.Button("Get Header")  # 조회 버튼
        get_output = gr.JSON(label="Output")  # 조회 결과를 표시할 출력 필드
        # 버튼 클릭 시 get_header_function 호출
        get_button.click(get_header_function, inputs=header_id_input, outputs=get_output)

# Gradio 앱 실행
# Gradio UI를 실행하여 웹에서 접근할 수 있도록 설정
demo.launch(share=True)
```

### Pydantic으로 데이터 검증
#### Custom Header, id는 정수만 서로 허락

```python
import sqlite3
import gradio as gr
from pydantic import BaseModel, ValidationError, conint

# 데이터베이스 파일 경로 설정
DATABASE_URL = "./test.db"  # SQLite 데이터베이스 파일 경로

# 데이터베이스 초기화 - 테이블 생성
# 데이터베이스 연결을 열고 테이블이 없을 경우 생성합니다.
with sqlite3.connect(DATABASE_URL) as conn:
    conn.execute('''
        CREATE TABLE IF NOT EXISTS header_info (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_agent TEXT,
            custom_header INTEGER
        )
    ''')

# 요청 모델 정의
class HeaderRequest(BaseModel):
    user_agent: str  # User-Agent는 문자열
    custom_header: conint(strict=True)  # Custom Header는 정수만 가능하도록 설정

# 요청 헤더 정보를 데이터베이스에 저장하는 함수
def store_header_function(user_agent: str, custom_header: str):
    # Pydantic 모델을 사용하여 입력 유효성 검사
    try:
        request_data = HeaderRequest(user_agent=user_agent, custom_header=int(custom_header))
    except ValidationError as e:
        return f"Validation error: {e}"

    # 데이터베이스에 연결하고 데이터를 저장
    with sqlite3.connect(DATABASE_URL) as conn:
        cursor = conn.cursor()
        # SQL 쿼리를 실행하여 header 정보를 삽입합니다.
        cursor.execute(
            "INSERT INTO header_info (user_agent, custom_header) VALUES (?, ?)",
            (request_data.user_agent, request_data.custom_header),
        )
        conn.commit()  # 변경사항을 커밋하여 데이터베이스에 반영
        return f"Header information stored successfully with id: {cursor.lastrowid}"

# 데이터베이스에서 헤더 정보를 가져오는 함수
def get_header_function(header_id: int):
    # 데이터베이스에 연결하고 특정 ID의 헤더 정보를 검색
    with sqlite3.connect(DATABASE_URL) as conn:
        cursor = conn.cursor()
        # SQL 쿼리를 실행하여 ID에 해당하는 header 정보를 조회합니다.
        cursor.execute("SELECT * FROM header_info WHERE id = ?", (header_id,))
        row = cursor.fetchone()  # 조회 결과에서 한 행을 가져옵니다.
        if row is None:
            return "Header not found"  # 해당 ID의 정보가 없을 경우 메시지 반환
        # 조회된 정보를 딕셔너리 형태로 반환
        return {
            "id": row[0],
            "user_agent": row[1],
            "custom_header": row[2],
        }

# Gradio 인터페이스 정의
with gr.Blocks() as demo:
    gr.Markdown("# Header Info Storage and Retrieval")  # 웹 UI 제목 표시
    # 헤더 정보를 저장하는 탭
    with gr.Tab("Store Header"):
        user_agent_input = gr.Textbox(label="User Agent")  # User-Agent 입력 필드
        custom_header_input = gr.Textbox(label="Custom Header (Integer Only)")  # Custom Header 입력 필드
        store_button = gr.Button("Store Header")  # 저장 버튼
        store_output = gr.Textbox(label="Output")  # 저장 결과를 표시할 출력 필드
        # 버튼 클릭 시 store_header_function 호출
        store_button.click(store_header_function, inputs=[user_agent_input, custom_header_input], outputs=store_output)
    # 헤더 정보를 조회하는 탭
    with gr.Tab("Get Header"):
        header_id_input = gr.Number(label="Header ID")  # 헤더 ID 입력 필드
        get_button = gr.Button("Get Header")  # 조회 버튼
        get_output = gr.JSON(label="Output")  # 조회 결과를 표시할 출력 필드
        # 버튼 클릭 시 get_header_function 호출
        get_button.click(get_header_function, inputs=header_id_input, outputs=get_output)

# Gradio 앱 실행
# Gradio UI를 실행하여 웹에서 접근할 수 있도록 설정
demo.launch(share=True)
```
