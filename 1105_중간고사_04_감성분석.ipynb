```python
import pandas as pd
df = pd.read_pickle(path).sample(frac=0.1)
df
```

```python
from konlpy.tag import Okt
import re
from tqdm import tqdm

okt = Okt()

stopwords = ['지난', '이하']  # Add more stopwords as needed

def extract_words_from_single_article(article): # Define a function for single articles
    clean_article = re.sub(r'[^\w\s]', '', str(article))

    nouns = okt.nouns(clean_article)

    filtered_nouns = [noun for noun in nouns if len(noun) > 1 and not re.match(r'^\d', noun) and noun not in stopwords]
    return filtered_nouns

df['추출단어'] = df['본문'].apply(extract_words_from_single_article)
# Join the list of words into a single string for each document
df['추출단어_str'] = df['추출단어'].apply(lambda x: ' '.join(x))
# 인덱스 생성
df.index = range(len(df))
df
```

### **Corpus 추출**

```python
# prompt: 추출단어_str 를 리스트 corpus로 만들기
# https://scikit-learn.org/1.5/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html --> corpus와 같은 형태
#
corpus = df['추출단어_str'].tolist()
corpus[:10]
```

### **<font color='orange'> Phase 5. LDA 분석단계**

### **감성분석**

```python
from transformers import pipeline

# 감성 분석 파이프라인 로드
sentiment_analyzer = pipeline("sentiment-analysis")

# 분석할 문장 리스트
sentences = [
    "I love this product! It's amazing!",
    "I'm not very happy with the service I received.",
    "The food was okay, but could be better.",
    "Absolutely terrible experience, I will never come back!",
    "Fantastic performance by the actors. Truly wonderful."
]

# 각 문장에 대한 감성 분석 수행
for sentence in sentences:
    result = sentiment_analyzer(sentence)
    label = result[0]['label']
    score = result[0]['score']
    print(f"Sentence: {sentence}\nSentiment: {label} (Score: {score:.2f})\n")
```

```python
# prompt: sentiment_analyzer의 토큰 수는

print(len(sentiment_analyzer.tokenizer))
```

```python
# prompt: df의 추출단어_str 10개만 감성분석하면

# 감성 분석 결과를 저장할 리스트 생성
sentiment_results = []

# df의 '추출단어_str' 컬럼에서 처음 10개의 문장에 대해 감성 분석 수행
for sentence in df['추출단어_str'][:10]:
    try:
        result = sentiment_analyzer(sentence)
        label = result[0]['label']
        score = result[0]['score']
        sentiment_results.append({'sentence': sentence, 'label': label, 'score': score})
    except Exception as e:
        print(f"Error analyzing sentence: {sentence}, Error: {e}")
        sentiment_results.append({'sentence': sentence, 'label': 'Error', 'score': 0})


# 결과 출력
for result in sentiment_results:
    print(f"Sentence: {result['sentence']}\nSentiment: {result['label']} (Score: {result['score']:.2f})\n")
```
