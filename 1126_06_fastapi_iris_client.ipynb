```python
# 클라이언트 코드 작성 (FastAPI 서버에 요청 보내기)
import requests

# 서버 URL 설정 (서버 실행 시 생성된 public_url을 사용하세요)
server_url = "https://01d9-35-196-92-24.ngrok-free.app/predict"  # YOUR_PUBLIC_URL을 실제 서버 URL로 변경하세요

# Iris 데이터 입력 (예: 꽃받침 길이, 꽃받침 너비, 꽃잎 길이, 꽃잎 너비)
iris_data = {
    "sepal_length": 5.1,
    "sepal_width": 3.5,
    "petal_length": 1.4,
    "petal_width": 0.2
}

# 서버로 POST 요청 보내기
response = requests.post(server_url, json=iris_data)

# 서버에서 반환된 예측 결과 출력
if response.status_code == 200:
    prediction = response.json().get("prediction")
    print(f"Predicted Iris class: {prediction}")
else:
    print(f"Failed to get prediction. Status code: {response.status_code}")
```

```python
def predict_iris(sepal_length, sepal_width, petal_length, petal_width):
    server_url = "https://01d9-35-196-92-24.ngrok-free.app/predict"  # Replace with your actual server URL
    iris_data = {
        "sepal_length": sepal_length,
        "sepal_width": sepal_width,
        "petal_length": petal_length,
        "petal_width": petal_width
    }
    response = requests.post(server_url, json=iris_data)

    if response.status_code == 200:
        prediction = response.json().get("prediction")
        class_mapping = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
        predicted_class = class_mapping.get(prediction)
        if predicted_class:
            return predicted_class
        else:
            return f"Invalid prediction value: {prediction}"
    else:
        return f"Failed to get prediction. Status code: {response.status_code}"

iface = gr.Interface(
    fn=predict_iris,
    inputs=[
        gr.Number(label="Sepal Length"),
        gr.Number(label="Sepal Width"),
        gr.Number(label="Petal Length"),
        gr.Number(label="Petal Width")
    ],
    outputs=gr.Textbox(label="Predicted Iris Class"),
    title="Iris Flower Prediction",
    description="Enter the sepal and petal measurements to predict the Iris flower class."
)

iface.launch()
```
