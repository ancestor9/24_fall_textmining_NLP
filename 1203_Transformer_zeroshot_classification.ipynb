### https://huggingface.co/facebook/bart-large-mnli

```python
from transformers import pipeline

# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(sequence_to_classify, candidate_labels)
#{'labels': ['travel', 'dancing', 'cooking'],
# 'scores': [0.9938651323318481, 0.0032737774308770895, 0.002861034357920289],
# 'sequence': 'one day I will see the world'}
```

### 데이터에 적용

```python
# Create a new column in the DataFrame for the genre predictions
def classify_lyrics(lyrics):
    result = classifier(lyrics, labels)  # lyrics의 내용을 분석하여 labels를 분류
    print(result)
    return result['labels'][0]  # Get the top label

tf['Predicted Genre'] = tf['Lyrics'].apply(classify_lyrics)
```

```python
from sklearn.datasets import fetch_20newsgroups

categories = ['rec.autos', 'talk.religion.misc', 'comp.graphics', 'sci.space', 'sci.med']

#학습 데이터셋을 가져옴
newsgroups_test = fetch_20newsgroups(subset='test',
                                     categories=categories)

print('#Test set size:', len(newsgroups_test.data))
print('#Selected categories:', newsgroups_test.target_names)
```

```python
df = pd.DataFrame(newsgroups_test.data, columns=['text'])
df['target'] = newsgroups_test.target
df
```

```python
# prompt: df의 target의  0,1,2,3에 해당되는 lable은
# Display the target labels and their corresponding categories
target_labels = newsgroups_test.target_names
for i, label in enumerate(target_labels):
  print(f"Target label {i}: {label}")
```

```python
# prompt: df['target']를 값을 label로 변경하라

df['target'] = df['target'].map({i: label for i, label in enumerate(target_labels)})
df
```

## **분류를 잘하는 방법은 chatGPT 혹은 데이터전처리**
### **1. 데이터 전처리 with HuggingFace**

```python
# prompt: tf['text']에서 From 뒹 사람이름, 이메일 주소 , 특수문자, 스페이스 등등 모두 제거해줘

import re

def clean_text(text):
    # Remove "From:" and everything before it
    text = re.sub(r'^.*?From:', '', text, flags=re.MULTILINE)

    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)

    # Remove special characters and spaces
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    return text

tf['text'] = tf['text'].apply(clean_text)
# 분류
tf['Predicted category'] = tf['text'].apply(classify_lyrics)
tf
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
import os

# OpenAI API 키 설정
from google.colab import userdata
OPENAI_API_KEY = userdata.get('openai-api-key')
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY

# GPT-4o-mini 모델 초기화
llm = ChatOpenAI(model="gpt-4o-mini",
                 temperature=0.5)

# 제로샷 분류를 위한 프롬프트 템플릿 생성
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that classifies text into given categories."),
    ("human", "Classify the following text into one of these categories: {labels}. Text: {text}")
])

# 체인 생성
chain = prompt | llm

# 분류 실행
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']

response = chain.invoke({
    "text": sequence_to_classify,
    "labels": ", ".join(candidate_labels)
})

print(response.content)
```
