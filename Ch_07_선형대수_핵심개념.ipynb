## **0.  2차원 공간의 두 행렬(벡터)의 내적**

```python
import numpy as np
import matplotlib.pyplot as plt

# 두 벡터 정의
a = np.array([3, 4])
b = np.array([4, -3])

# 내적 계산
dot_product = np.dot(a, b)

# 벡터 시각화
fig, ax = plt.subplots()
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')

# 그래프 설정
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Dot Product: {dot_product}')
plt.legend()
plt.show()
```

## **1.  2차원 공간의 두 벡터의 선형결합**

### 두 벡터에 스칼라를 곱하고 합한 결과
두 벡터 \\( \mathbf{v_1} = \begin{bmatrix} v_{11} \\ v_{12} \end{bmatrix} \\)과 \\( \mathbf{v_2} = \begin{bmatrix} v_{21} \\ v_{22} \end{bmatrix} \\)에 각각 스칼라 \\( c_1 \\), \\( c_2 \\)를 곱한 후 합한 결과는 다음과 같이 정의됩니다:

$$ c_1 \mathbf{v_1} + c_2 \mathbf{v_2} = \begin{bmatrix} c_1 \cdot v_{11} + c_2 \cdot v_{21} \\ c_1 \cdot v_{12} + c_2 \cdot v_{22} \end{bmatrix} $$

예를 들어, 벡터 \\( \mathbf{v_1} = \begin{bmatrix} 2 \\ 3 \end{bmatrix} \\), \\( \mathbf{v_2} = \begin{bmatrix} 1 \\ 4 \end{bmatrix} \\), 스칼라 \\( c_1 = 2 \\), \\( c_2 = 3 \\)일 때, 다음과 같이 계산됩니다:

$$ 2 \mathbf{v_1} + 3 \mathbf{v_2} = \begin{bmatrix} 2 \cdot 2 + 3 \cdot 1 \\ 2 \cdot 3 + 3 \cdot 4 \end{bmatrix} = \begin{bmatrix} 7 \\ 18 \end{bmatrix} $$



```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터와 스칼라 정의
v1 = np.array([2, 3])
v2 = np.array([1, 4])
c1 = 2
c2 = 3

# 스칼라 곱 후 결과
v1_scaled = c1 * v1
v2_scaled = c2 * v2

# 스칼라 곱 결과의 합
v_result = v1_scaled + v2_scaled

# 시각화
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', linestyle='-', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', linestyle='-', label='v2')
plt.quiver(0, 0, v1_scaled[0], v1_scaled[1], angles='xy', scale_units='xy', scale=1, color='g', linestyle='-', label='c1 * v1 (c1=2)')
plt.quiver(0, 0, v2_scaled[0], v2_scaled[1], angles='xy', scale_units='xy', scale=1, color='purple', linestyle='-', label='c2 * v2 (c2=3)')
plt.quiver(0, 0, v_result[0], v_result[1], angles='xy', scale_units='xy', scale=1, color='orange', linestyle='-', label='c1 * v1 + c2 * v2')

# 축 범위 설정
plt.xlim(0, 15)
plt.ylim(0, 20)
```

## **2. Span과 선형 결합의 개념**

- **Span**: Span은 주어진 벡터들의 모든 선형 결합으로 만들어지는 공간을 의미합니다. 예를 들어, 2차원 벡터 x와 y가 있다고 할 때, 이 두 벡터의 Span은 두 벡터의 선형 결합을 통해 형성되는 평면상의 모든 점들을 나타냅니다. 즉, 벡터 x와 y가 생성하는 공간입니다. 만약 두 벡터가 선형 독립적이라면, 그 Span은 2차원 평면 전체가 됩니다.

- **선형 결합(Linear Combination)**: 선형 결합은 여러 벡터의 스칼라 배수를 더하는 것을 의미합니다. 예를 들어, 벡터 x와 y가 있을 때, 이들의 선형 결합은 c * x + d * y의 형태로 나타낼 수 있습니다. 여기서 c와 d는 임의의 스칼라 값입니다. 선형 결합을 통해 우리는 기존 벡터들로부터 새로운 벡터를 만들어낼 수 있으며, 이를 통해 벡터들이 만들어내는 전체 공간(Span)을 정의할 수 있습니다.

### 요약

아래 코드는 두 벡터 x와 y의 선형 결합을 시각화하여 선형 대수에서 중요한 개념인 벡터의 결합 가능성을 시각적으로 보여줍니다. 각 벡터는 스칼라 계수의 조합에 따라 다른 방향과 크기를 가지며, 그래프에서 이 모든 결합을 보여줌으로써 벡터 공간을 이해하는 데 도움을 줍니다. Span 개념을 통해 이 벡터들이 만들 수 있는 전체 공간을 이해할 수 있으며, 선형 결합은 이러한 벡터들의 조합 방법을 설명합니다.

### **2차원 공간의 Span**
두 벡터 \\( \mathbf{v_1} = \begin{bmatrix} v_{11} \\ v_{12} \end{bmatrix} \\)과 \\( \mathbf{v_2} = \begin{bmatrix} v_{21} \\ v_{22} \end{bmatrix} \\)의 span은 다음과 같이 정의됩니다:

$$ \text{span}(\mathbf{v_1}, \mathbf{v_2}) = \{ c_1 \mathbf{v_1} + c_2 \mathbf{v_2} \mid c_1, c_2 \in \mathbb{R} \} $$

### 선형결합
벡터 \\( \mathbf{v_1} \\)과 \\( \mathbf{v_2} \\)의 선형결합은 다음과 같습니다:

$$ \mathbf{v} = c_1 \mathbf{v_1} + c_2 \mathbf{v_2} $$

여기서 \\( c_1 \\)과 \\( c_2 \\)는 스칼라 계수이고, \\( \mathbf{v_1} \\)과 \\( \mathbf{v_2} \\)는 벡터입니다.

### 행렬 표현
행렬 형태로 두 벡터 \\( \mathbf{v_1}, \mathbf{v_2} \\)의 선형결합을 표현하면 다음과 같습니다:

$$ \mathbf{A} \mathbf{c} =
\begin{bmatrix}
| & | \\
\mathbf{v_1} & \mathbf{v_2} \\
| & |
\end{bmatrix}
\begin{bmatrix}
c_1 \\
c_2
\end{bmatrix} =
\begin{bmatrix}
v_{11} & v_{21} \\
v_{12} & v_{22}
\end{bmatrix}
\begin{bmatrix}
c_1 \\
c_2
\end{bmatrix} $$

여기서 \\( \mathbf{A} \\)는 열벡터 \\( \mathbf{v_1}, \mathbf{v_2} \\)로 구성된 행렬입니다.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define two 2D vectors
x = np.array([2, 1])
y = np.array([-1, 3])

# Define scalar coefficients for linear combinations
coefficients = np.linspace(-2, 2, 10)

# Calculate linear combinations
combinations = np.array([c * x + d * y for c in coefficients for d in coefficients])

# Plot the original vectors
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# Plot linear combinations
for i in range(len(combinations)):
    plt.quiver(0, 0, combinations[i, 0], combinations[i, 1],
               angles='xy', scale_units='xy', scale=1, color='g', alpha=0.3)

# Set plot limits and properties
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.legend()
plt.title('Linear Combinations of 2D Vectors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

## **3. 역 행렬(Inverse Matrix)**

### 역행렬 (Inverse Matrix)
정사각행렬 \\( \mathbf{A} \\)의 역행렬 \\( \mathbf{A}^{-1} \\)은 다음과 같은 조건을 만족하는 행렬입니다:

$$ \mathbf{A} \cdot \mathbf{A}^{-1} = \mathbf{I} $$

여기서 \\( \mathbf{I} \\)는 단위행렬 (Identity Matrix)입니다. \\( \mathbf{A} \\)의 역행렬이 존재하려면 \\( \mathbf{A} \\)는 정칙행렬 (즉, 행렬식이 0이 아닌 행렬)이어야 합니다.

예를 들어, 행렬 \\( \mathbf{A} \\)가 다음과 같다고 가정합니다:

$$ \mathbf{A} = \begin{bmatrix} 2 & 1 \\ 7 & 4 \end{bmatrix} $$

이 행렬의 역행렬 \\( \mathbf{A}^{-1} \\)은 다음과 같이 계산할 수 있습니다.


```python
A = np.array([[2, 1],
              [7, 4]])

# 역행렬 계산
try:
    A_inv = np.linalg.inv(A)
    print("행렬 A의 역행렬:\n", A_inv)
except np.linalg.LinAlgError:
    print("역행렬이 존재하지 않습니다. 행렬이 정칙이 아닙니다.")

# 결과 확인
# A와 A_inv를 곱해서 단위 행렬이 나오는지 확인
I = np.dot(A, A_inv)
print("\nA와 A의 역행렬의 곱:\n", I)
```

## **4. 선형변환(Linear Transformation)**

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageFilter
import urllib

# 이미지 불러오기
url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
image = Image.open(urllib.request.urlopen(url))

# shear_factor 값 리스트 생성
shear_factors = np.linspace(0.5, 0.9, 5)

# 원본 이미지와 변환된 이미지를 한 줄에 표시
plt.figure(figsize=(20, 5))

# 원본 이미지 시각화
plt.subplot(1, 6, 1)
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')

# shear 변환 적용 및 시각화
for i, shear_factor in enumerate(shear_factors):
  shear_matrix = (1, shear_factor, 0, 0, 1, 0)
  sheared_image = image.transform(image.size, Image.AFFINE, shear_matrix, Image.BICUBIC)
  plt.subplot(1, 6, i + 2)
  plt.title(f"Shear factor: {shear_factor:.2f}")
  plt.imshow(sheared_image)
  plt.axis('off')

plt.show()
```
