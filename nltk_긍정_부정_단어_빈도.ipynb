```python
import nltk
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

# NLTK 데이터 다운로드 (처음 실행 시)
nltk.download('movie_reviews')
nltk.download('stopwords')

# 불용어 리스트
stop_words = set(stopwords.words('english'))

# 긍정적 리뷰와 부정적 리뷰의 파일 ID를 분리
positive_ids = movie_reviews.fileids('pos')
negative_ids = movie_reviews.fileids('neg')

def get_word_frequencies(ids):
    words = []
    for file_id in ids:
        words.extend(movie_reviews.words(file_id))

    # 불용어 제거 및 소문자 변환
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]
    return Counter(filtered_words)

# 긍정적 리뷰와 부정적 리뷰의 단어 빈도 계산
positive_freq = get_word_frequencies(positive_ids)
negative_freq = get_word_frequencies(negative_ids)

# 상위 10개의 단어 시각화
def plot_word_frequencies(freq, title):
    common_words = freq.most_common(10)
    words, counts = zip(*common_words)
    plt.bar(words, counts)
    plt.title(title)
    plt.show()

plot_word_frequencies(positive_freq, 'Top 10 Words in Positive Reviews')
plot_word_frequencies(negative_freq, 'Top 10 Words in Negative Reviews')
```
